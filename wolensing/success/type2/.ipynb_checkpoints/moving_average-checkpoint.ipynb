```python
import numpy as np
```

```python
ft = np.loadtxt('./5m45_ft.txt')
ts = np.loadtxt('./5m45_ts.txt')
```

```python
import matplotlib.pyplot as plt
plt.plot(ts, ft)
```

```python
def moving_average(data, window_size):
    """
    Calculate the moving average of a 1D array.

    Parameters:
    - data: list or numpy array of numbers.
    - window_size: size of the moving window.

    Returns:
    - numpy array containing the moving averages.
    """
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')

window = 100
smooth_Ft = moving_average(ft, window)
x_trimmed = ts[window - 1:]
plt.plot(x_trimmed, smooth_Ft)
plt.xlim(1.685e-5, 1.7e-5)
```
