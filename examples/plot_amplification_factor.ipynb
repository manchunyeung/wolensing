```python
import wolensing.amplification_factor.amplification_factor as af
import numpy as np
```

### Import the saved amplification factor F(t) and F(f)

```python
scale = np.sqrt(9)
# ts = np.loadtxt('./data/test/test_sis_ts.txt')
# F_tilde = np.loadtxt('./data/test/test_sis_F_tilde.txt')
ws = np.loadtxt('./data/t2_50ws.txt')
Fws = np.loadtxt('./data/t2_50Fws.txt', dtype=complex, converters={0: lambda s: complex(s.decode().replace('+-', '-'))})/scale
```

### Import the data to the class

```python

kwargs = {'mu':1}
amplification = af.amplification_factor(kwargs_macro=kwargs)
amplification.importor(freq=True, fs=ws, Fws=Fws)
# amplification.importor(time=True, ts=ts, F_tilde=F_tilde) # both time and freq can be input at the same time
```

### Plot F(t)

```python
amplification.plot_time()
```

### Plot the absolute value of F(f)

```python
amplification.plot_freq() # default to plot the absolute value
```

### Plot the phase of F(f)

```python
amplification.plot_freq(pha=True)
```
